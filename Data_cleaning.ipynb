```python
# Show trees with 'status' == 'Alive'
# Replace 'NaN' in spc_latin with '(Unknown)'
# Get 25% of 'tree_dbh' for each 'spc_latin' and apply them to values smaller than 25% 
# Get 75% of 'tree_dbh' for each 'spc_latin' and apply them to values bigger than 75%
# The following columns are requires: 'tree_id', 'tree_dbh', 'stump_diam', 'curb_loc', 'status', 'health', 'spc_latin', 'steward'
# Merge the list of 'Alive' trees with 'describe' stats of grouped alive tree lists by 'spc_latin' with the values of '25%' and '75%'
```

```python
import pandas as pd
```

```python
tree_census = pd.read_csv('2015_Street_Tree_Census_-_Tree_Data_20250109.csv')
tree_census
```

```python
tree_census.columns
```

```python
tree_census_subset = tree_census[['tree_id', 'tree_dbh', 'stump_diam', 
        'curb_loc', 'status', 'health', 'spc_latin', 'steward']]
tree_census_subset_alive = tree_census_subset[tree_census_subset['status'] == 'Alive']
tree_census_subset_alive
```

```python
tree_census_subset_alive.isna().sum()
```

```python
mask = tree_census_subset_alive['spc_latin'].isna()
tree_census_subset_alive.loc[mask, 'spc_latin'] = tree_census_subset_alive.loc[mask, 'spc_latin'].fillna('(Unknown)')
tree_census_subset_alive.groupby('spc_latin').describe()
```

```python
stats_alive = tree_census_subset_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin','25%','75%']]
stats_alive
```

```python
merged_tree_census_subset_alive = tree_census_subset_alive.merge(stats_alive, on='spc_latin', how='left')
merged_tree_census_subset_alive
```

```python
mask = merged_tree_census_subset_alive['tree_dbh'] < merged_tree_census_subset_alive['25%']
merged_tree_census_subset_alive.loc[mask, 'tree_dbh'] = merged_tree_census_subset_alive['25%']

mask = merged_tree_census_subset_alive['tree_dbh'] > merged_tree_census_subset_alive['75%']
merged_tree_census_subset_alive.loc[mask, 'tree_dbh'] = merged_tree_census_subset_alive['75%']

merged_tree_census_subset_alive
```
